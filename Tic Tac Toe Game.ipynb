```python
from IPython.display import clear_output
def show(value):
    row1=''
    row2=''
    row3=''
    for item in range(0,3):
        row1 = row1 + '|' + value[item]
    row1=row1+'|'
    print(row1)
    
    for item in range(3,6):
        row2=row2+'|' + value[item]
    row2 = row2 + '|'
    print(row2)
    
    for item in range(6,9):
        row3 = row3 + '|' + value[item]
    row3 = row3 + '|'
    print(row3)

def check_win(numbers):
    value_x = []
    value_o = []
    for i in range(0, len(numbers),2):
        value_x.append(numbers[i])
    for i in range(1,len(numbers),2):
        value_o.append(numbers[i])
    na = [[1,2,3] , [1,4,7] , [1,5,9] , [4,5,6] ,[7,8,9] , [7,5,3] , [7,4,1] , [8,5,2] ,[9,5,1] ,[9,8,7] ,[6,5,4],[3,2,1],[3,5,7],[2,5,8] ,[3,6,9]]
    if len(value_x) >= 3:
        if value_x[0:3] in na:
            return 'x'
    if len(value_x) >= 4:
        if value_x[1:4] in na:
            return 'x'
    if len(value_x) ==5 :
        if value_x[2:5] in na:
            return 'x'
    if len(value_o) == 3:
        if value_o[0:3] in na:
            return 'o'
    if len(value_o) >= 4:
        if value_o[1:3] in na:
            return 'o'
    return 'd'

def validate(list):
    for item in list:
        if list.count(item) > 1:
            list.pop(list.index(item))

def resend():
    value = input("Do You want to play again input y OR N: ")
    while value not in ['y','Y','n','N']:
        value = input("Do you want to play again input Y or N: ")
    if value.capitalize() == 'Y':
        print("Game has restarted")
        game()
    if value.capitalize() == 'N':
        print("Thank you  for playing the game")
    

def game():
    holder = ['1','2','3','4','5','6','7','8','9']
    show(holder)
    stopper = False
    count = 0
    checker = []
    place_holder = 0
    print("Welcome to Toriola Tic tac toe")
    for i in range(0,6):
        while stopper == False:
            hold = input('Player x please select your position')
            if hold in ['1','2','3','4','5','6','7','8','9']:
                place_holder = int(hold)
                if place_holder in checker: 
                    while place_holder in checker:
                        hold = input('The position you have selected is occupied please select available spaces: ')
                        place_holder = int(hold)
                    checker.append(place_holder)
                    holder[place_holder-1] = 'x'
                checker.append(place_holder)
                validate(checker)
                holder[place_holder-1] = 'x'
                clear_output()
                if len(checker) >= 5:
                    if check_win(checker) == 'x' or check_win(checker) == 'o':
                        print(f"Player {check_win(checker)} wins")
                        resend()
                        stopper = True
                        pass
                    elif check_win(checker) == 'd' and len(checker) == 9:
                        print("The game is a draw")
                        resend()
                print(checker)
                show(holder)
            elif hold not in ['1','2','3','4','5','6','7','8','9']:
                print("Input correct place holders")
                break
            count=count+1
            if count == 5:
                stopper = True
            hold = input('Player o Please select your positon')
            if hold in ['1','2','3','4','5','6','7','8','9']:
                place_holder = int(hold)
                if place_holder in checker: 
                    while place_holder in checker:
                        hold = input('The position you have selected is occupied please select available spaces: ')
                        place_holder = int(hold)
                    checker.append(place_holder)
                    holder[place_holder-1] = 'o'
                checker.append(place_holder)
                validate(checker)
                holder[place_holder-1] = 'o'
                clear_output()
                if len(checker) >= 6:
                    if check_win(checker) == 'x' or check_win(checker) == 'o':
                        print(f"Player {check_win(checker)} wins")
                        resend()
                        stopper = True
                    elif check_win(checker) == 'd' and len(checker) == 9:
                        print("The game is a draw")
                        resend()
                show(holder)
            elif hold not in ['1','2','3','4','5','6','7','8','9']:
                print("incorect place holder")
                break
                
game()
    
    
        
        
```
